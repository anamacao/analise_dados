```python
import numpy as np
import pandas as pd
```

```python
#seleção (filtragem de linhas específicas de um dataframe)
dataframe_01=pd.DataFrame({
    "região":["america do sul", "america do sul", "america do sul", "america do sul","africa","asia"],
    "estados":["brasil","argentina","uruguai","paraguai","angola","timor leste"],
    "idh":[0.754,0.842,0.809,0.724,0.574, 0.607]
                           })
dataframe_01
```

```python
região=dataframe_01["região"]=="asia"
região
```

```python
dataframe_01[região]
```

```python
região=dataframe_01[dataframe_01["região"]=="asia"]
região
```

```python
busca=dataframe_01["região"]!="africa"
dataframe_01[busca]
```

```python
busca=dataframe_01["idh"]<0.809
dataframe_01[busca]
```
